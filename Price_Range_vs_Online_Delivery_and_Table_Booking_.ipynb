```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load Dataset
df = pd.read_csv("Dataset .csv")
```

```python
# Step 3: Basic Inspection
print(df.columns)
print(df.head())
```

```python
# Step 4: Check important columns
# Should have: 'Price range', 'Has Online delivery', 'Has Table booking'

# Step 5: Analyze relationship between Price Range and Online Delivery
price_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'])

print("\nPrice Range vs Online Delivery:")
print(price_delivery)

# Plot Online Delivery
price_delivery.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Price Range vs Online Delivery Availability')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.grid(True)
plt.legend(title='Online Delivery')
plt.show()
```

```python
# Step 6: Analyze relationship between Price Range and Table Booking
price_table = pd.crosstab(df['Price range'], df['Has Table booking'])

print("\nPrice Range vs Table Booking:")
print(price_table)

# Plot Table Booking
price_table.plot(kind='bar', stacked=True, figsize=(10,6), colormap='plasma')
plt.title('Price Range vs Table Booking Availability')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.grid(True)
plt.legend(title='Table Booking')
plt.show()
```

```python
# Step 7: Calculate proportions (Optional Deep Dive)
# Find % of restaurants in each price range offering delivery/booking
delivery_percentage = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
table_percentage = df.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack() * 100

print("\nDelivery Availability Percentage by Price Range:")
print(delivery_percentage)

print("\nTable Booking Availability Percentage by Price Range:")
print(table_percentage)
```
